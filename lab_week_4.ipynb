```python
"""
Lab Tasks - Depth First Search (DFS)
Romania Travelling Salesman Problem: Arad → Bucharest
"""

from collections import defaultdict, deque


romania_graph = {
    'Arad':           [('Zerind', 75),   ('Sibiu', 140),  ('Timisoara', 118)],
    'Zerind':         [('Arad', 75),     ('Oradea', 71)],
    'Oradea':         [('Zerind', 71),   ('Sibiu', 151)],
    'Sibiu':          [('Arad', 140),    ('Oradea', 151), ('Fagaras', 99),   ('Rimnicu Vilcea', 80)],
    'Timisoara':      [('Arad', 118),    ('Lugoj', 111)],
    'Lugoj':          [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia':        [('Lugoj', 70),    ('Drobeta', 75)],
    'Drobeta':        [('Mehadia', 75),  ('Craiova', 120)],
    'Craiova':        [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Rimnicu Vilcea': [('Sibiu', 80),    ('Craiova', 146), ('Pitesti', 97)],
    'Fagaras':        [('Sibiu', 99),    ('Bucharest', 211)],
    'Pitesti':        [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Bucharest':      [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90),  ('Urziceni', 85)],
    'Giurgiu':        [('Bucharest', 90)],
    'Urziceni':       [('Bucharest', 85), ('Vaslui', 142),  ('Hirsova', 98)],
    'Hirsova':        [('Urziceni', 98),  ('Eforie', 86)],
    'Eforie':         [('Hirsova', 86)],
    'Vaslui':         [('Urziceni', 142), ('Iasi', 92)],
    'Iasi':           [('Vaslui', 92),    ('Neamt', 87)],
    'Neamt':          [('Iasi', 87)],
}

─
def separator(char='─', width=60):
    print(char * width)




def dfs(graph, start, goal):
    """
    Iterative DFS using an explicit stack.
    Returns (path, total_distance, nodes_visited_order).
    """

    stack = [(start, [start], 0)]
    visited = set()
    nodes_visited = []

    while stack:
        node, path, cost = stack.pop()

        if node in visited:
            continue
        visited.add(node)
        nodes_visited.append(node)

        if node == goal:
            return path, cost, nodes_visited


        for neighbour, weight in reversed(graph[node]):
            if neighbour not in visited:
                stack.append((neighbour, path + [neighbour], cost + weight))

    return None, float('inf'), nodes_visited   # goal not reachable


separator('═')
print("  LAB TASK 1 – DFS: Arad → Bucharest")
separator('═')

path, distance, order = dfs(romania_graph, 'Arad', 'Bucharest')

print(f"\nNodes visited (DFS order): {' → '.join(order)}")
print(f"\nPath found   : {' → '.join(path)}")
print(f"Total distance: {distance} km")
separator()




def bfs(graph, start, goal):
    """
    BFS using a deque.
    Returns (path, total_distance, nodes_visited_order).
    """
    queue = deque([(start, [start], 0)])
    visited = set()
    nodes_visited = []

    while queue:
        node, path, cost = queue.popleft()

        if node in visited:
            continue
        visited.add(node)
        nodes_visited.append(node)

        if node == goal:
            return path, cost, nodes_visited

        for neighbour, weight in graph[node]:
            if neighbour not in visited:
                queue.append((neighbour, path + [neighbour], cost + weight))

    return None, float('inf'), nodes_visited


def compare(start, goal, label):
    separator('─')
    print(f"  {label}")
    print(f"  Start: {start}  |  Goal: {goal}")
    separator('─')

    bfs_path, bfs_dist, bfs_order = bfs(romania_graph, start, goal)
    dfs_path, dfs_dist, dfs_order = dfs(romania_graph, start, goal)

    print(f"\nBFS  nodes visited ({len(bfs_order):2d}): {' → '.join(bfs_order)}")
    print(f"BFS  path : {' → '.join(bfs_path)}  [{bfs_dist} km]")

    print(f"\nDFS  nodes visited ({len(dfs_order):2d}): {' → '.join(dfs_order)}")
    print(f"DFS  path : {' → '.join(dfs_path)}  [{dfs_dist} km]")

    if len(bfs_order) < len(dfs_order):
        winner = "BFS visits FEWER nodes ✓"
    elif len(dfs_order) < len(bfs_order):
        winner = "DFS visits FEWER nodes ✓"
    else:
        winner = "Both visit the SAME number of nodes"

    print(f"\n  → {winner}  (BFS={len(bfs_order)}, DFS={len(dfs_order)})")
    separator()


separator('═')
print("  LAB TASK 2 – BFS vs DFS Node-Visit Comparison")
separator('═')

compare('Arad', 'Timisoara',
        "Case A: BFS visits FEWER nodes than DFS")


        "Case B: DFS visits FEWER nodes than BFS")
```
