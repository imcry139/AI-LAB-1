```python
def alphabeta(node, depth, alpha, beta, is_max):
    # Base case (leaf node)
    if depth == 0 or isinstance(node, int):
        return node

    if is_max:
        max_eval = float('-inf')
        for child in node:
            eval = alphabeta(child, depth - 1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)

            # Pruning condition
            if beta <= alpha:
                break

        return max_eval

    else:
        min_eval = float('inf')
        for child in node:
            eval = alphabeta(child, depth - 1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)

            # Pruning condition
            if beta <= alpha:
                break

        return min_eval

```

```python
### minmax


def minimax(node, depth, is_max):
    # If leaf node
    if depth == 0 or isinstance(node, int):
        return node

    if is_max:
        best = float('-inf')
        for child in node:
            val = minimax(child, depth - 1, False)
            best = max(best, val)
        return best
    else:
        best = float('inf')
        for child in node:
            val = minimax(child, depth - 1, True)
            best = min(best, val)
        return best


# Tree structure based on your diagram
tree = [
    [  # M (MIN)
        [-4, -5],  # Q (MAX)
        [2, 3]     # R (MAX)
    ],
    [  # N (MIN)
        [4, 5],    # O (MAX)
        [6, 7]     # P (MAX)
    ]
]

result = minimax(tree, 3, True)
print("Optimal Value:", result)
```
